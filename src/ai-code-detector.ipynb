# Detection of AI-generated code

A multi-signal classifier that estimates probability of AI-generated code

code -> embedding -> classifier -> AI vs human


Combine:
- Text embeddings (transformer)
- Code structure features (AST)
- Statistical signals (perplexity, entropy)

Later:
- Adversarial training
- Detector vs generator game
- Stylometry + watermarking detection

Data Gathering
- AI code via LLM api requests
- Human code from github
